Objective: Distance 

Minimize Distance of the route, made of legs. 

Min SUM of {i=1:13} {j=1:13} c[i,j] * x[i,j]

x[i,j] is a binary integer. If its 0, i->j is not part of the plan; if its 1, means we're going from i->j. 

x[i,j] = {1 0}

Constraints:

* Sales person can only be in one city at a time.
* Cannot travel from $ c_i -> c_i $.
* Cannot travel back to the city last traveled from: $ c_i -> c_j -> c_i $
* When i=j, c=0
* Route must include every city.
* Route must start and end at the depot 0 (New York).

1. x[i,i] = 0 for all i. (for loop constraint)

2. row sums equal one - can only go to each city once. city i cannot be in city 2 as well. sum of x[i,1] + x[i,2] + ... + x[i,13] = 1 for every i ... cannot go from 1 to 2 and 3.

3. col sums equal one - must go to each city once; th x[1,k] + x[2,j] + ... + x[13,j] = 1 for every i ,.... cannot go from 1 and 3 to 2. 

4. dont want to go to a city you already been to? (e.g. must visit every city , cant do loops ... )
